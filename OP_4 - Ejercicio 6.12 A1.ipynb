# Ejercicio 6.12 (A1)

Encuentre todos los eigenvalores y eigenvectores de las siguientes matrices:

$$ A_1 = \begin{bmatrix} 1 & 1 \\ 0 & i \end{bmatrix} $$

### Paso 1: Sistema de ecuaciones representativo
Partimos de la relación de transformada:
$$ Av = y $$
Donde $v = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} $ y $ y = \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} $

De esta relación podemos extraer un sistema de ecuaciones:

$\begin{bmatrix} 1 & 1 \\ 0 & i \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} y_1 \\ y_2 \end{bmatrix}$


$ \begin{cases} x_1 + x_2 = y_1 \\ ix_2 = y_2 \end{cases} $

### Paso 2: Construyendo la ecuación característica

Nuestra meta es encontrar **eigenvalores** $\lambda$ y **eigenvectores** $v$ de manera que:
$$ y = \lambda v $$
Esto es, buscar valores de $\lambda$ tales que:
$$  x_1 + x_2 = \lambda x_1 \\ ix_2 = \lambda x_2  $$


Reordenando tenemos:
$  (1- \lambda)x_1 + x_2 = 0 \\ (i-\lambda)x_2 = 0 \tag 1 $ 

Esto puede ser escrito utilizando la notación matricial de la siguiente manera:
$$ A - \lambda I = 0 $$
O bien,
$$ \left( A - \begin{bmatrix} \lambda & 0 \\ 0 & \lambda \end{bmatrix} \right)v =0 $$

### Paso 3: Resolviendo el polinomio caracteristico


Para encontrar los valores de $\lambda$, resolvemos $|A-\lambda I|=0$, lo que es denominada la ecuación caracteristica.


$$ A - \lambda I = \begin{vmatrix} 1-\lambda & 1 \\ 0 & i-\lambda \end{vmatrix}  $$
$$ = (1-\lambda)(i-\lambda) - (0) (1) $$
$$ = (1-\lambda)(i-\lambda) $$

Ahora, esto equivale a 0, por lo tanto:
$$(1-\lambda)(i-\lambda) =0$$
Lo que significa que los eigenvalores correspondientes son $ \lambda_1 = 1 $ y $\lambda_2 = i $

### Paso 4: Aplicando eigenvalores a las ecuaciones dadas por $|A - \lambda I|$
**Caso 1:** $\lambda = 1$
$(1- 1)x_1 + x_2 = 0 \\ (i-1)x_2 = 0 $

Obtenemos un sistema de la siguiente forma:
$\begin{pmatrix} 0 & 1 \\ 0 & i-1 \end{pmatrix}$, luego, procedemos aplicando eliminación gausiana ($R_2 = R_2 + (-i+1)R_1$) obtenemos: $\begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}$, y de aquí deducimos que:
$$ x_2 = 0 , x_1 = x_1 $$

El conjunto solución es: $x = \begin{pmatrix} x_1 \\ 0 \end{pmatrix}$, tomando arbitrariamente $x_1 = 1$, obtenemos el **eigenvector**:
$$v_1 =  \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$

**Caso 2:** $\lambda = i$
$(1- i)x_1 + x_2 = 0 \\ (i-i)x_2 = 0 $

Obtenemos un sistema de la siguiente forma:
$\begin{pmatrix} 1-i & 1 \\ 0 & 0 \end{pmatrix}$, luego, tranformamos la matriz de manera que $R_1 = \frac{R_1}{1-i} $ obtenemos: $\begin{pmatrix} 1 & \frac{1+i}{2} \\ 0 & 0 \end{pmatrix}$, y de aquí deducimos que:
$$ x_1 = -(1+i)x_2, x_2 = x_2 $$

El conjunto solución es: $ \left\{ x_2 \begin{pmatrix} -\frac{(1+i)}{2} \\ 1 \end{pmatrix} \right\} $, tomando arbitrariamente $x_2 = 1$, obtenemos el **eigenvector**:
$$v_2 =  \begin{pmatrix} \frac{-1-i}{2} \\ 1 \end{pmatrix}$$

## Solución numérica
Habiendo comprendido la escencia de la obtención de eigenvalores y eigenvectores, podemos exponer una manera alternativa de atacar este problema.
