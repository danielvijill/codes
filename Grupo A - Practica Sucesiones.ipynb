#### 1). 
$$\left\{\dfrac{n^2}{n+5}\right\} \: n\in \mathbb{N}$$

```python
from sympy import init_printing
init_printing(use_latex=True)
from sympy import *
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import pyplot
import ipywidgets as wg
from IPython.display import display
import numpy as np
```

```python
def suc(n):
    if n >=0:
        return (n**2)/(n+5)
    else:
        print("'n' no es natural")
```

```python
suc0=[suc(0)]
suc1=[suc(1)]
suc2=[suc(2)]
suc3=[suc(3)]
suc4=[suc(4)]
suc0,suc1,suc2,suc3,suc4
```

```python
suc5=[suc(5)]
suc6=[suc(6)]
suc7=[suc(7)]
suc8=[suc(8)]
suc9=[suc(9)]
suc5,suc6,suc7,suc8,suc9
```

```python
suc10=[suc(10)]
suc11=[suc(11)]
suc12=[suc(12)]
suc13=[suc(13)]
suc14=[suc(14)]
suc100=[suc(100)]
suc10,suc11,suc12,suc13,suc14,suc100
```

```python
n = np.linspace(0, 10, 10)
y = (n**2)/(n+5)
plt.plot(n ,y ,"b.")
```

```python
n = np.linspace(0, 50, 50)
y = (n**2)/(n+5)
plt.plot(n ,y ,"g.")
```

```python
n = np.linspace(0, 100, 50)
y = (n**2)/(n+5)
plt.plot(n ,y ,"c.")
```

**Se observa que la sucesión Diverge.**

```python
def sucesion(n):
    
    n1=[]
    y=[]
    for i in range(n):
        n1=n1+[i]
        y=y+[suc(i)]
    plt.plot(n1 ,y ,"b.")
```

```python
n_slide=wg.IntSlider(value=1,min=0,max=100, step=1)
wg.interact(sucesion,n=n_slide)
```

#### 1). $$ f_n(x)=\left\{ \begin{array}1x^n & 0\leq x \leq 1 \\ 1  & x\geq1 \end{array}\right. $$
