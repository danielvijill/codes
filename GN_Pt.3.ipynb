### Ejemplo 2.1
Para la función constante $f(x) = 2$ del espacio $C_{[−1,1]}$, se tiene que la norma $||f||_{1} = 4.$

```python
from matplotlib import pyplot

def f(x):
    return 2

x = range(-10, 10)


pyplot.plot(x, [f(i) for i in x])


pyplot.axhline(0, color="red")
pyplot.axvline(0, color="red")


plt.title("grafico de la función")
pyplot.xlim(-1, 1)
pyplot.ylim(0, 10)

pyplot.show()
```

La región encerrada por el gráfico
de $f$, las rectas$ x = −1$, $x = 1$ y el eje
$X$ es un rectángulo de base $2$ y altura
$2$, el área de la región acotada es $4$

$\therefore$ $||f||_{1} = 4$

### Ejemplo 2.2
Para la función constante $f(x) = \frac{x}{2}$ del espacio $C_{[0,2]}$, se tiene que la norma $||f||_{1} = 1.$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
x = np.linspace(0, 1, 100)


plt.plot(x, x/2, label='x/2')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title("grafico de la función")
plt.legend()
plt.show()
```

La región acotada por el gráﬁco de $f$, las rectas $x = 0, x = 2$ y el eje de las abscisas es un triángulo de base $1$ y altura $2$, el área de la región acotada es $\frac{(2)(1)}{2}= 1.$

$\therefore ||f||_{1} = 1$

### Ejemplo 2.3
Para la función constante $f(x) =\sqrt{1-x^{2}}$  del espacio $C_{[-1,1]}$, se tiene que la norma $||f||_{1} = \frac{\pi}{2}.$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 200)
plt.axhline(0, color="red")
plt.axvline(0, color="red")

plt.plot(x, np.sqrt(1-x**2), label='f(x)')

plt.title("grafico de la función")
plt.legend()
plt.show()
```

La región determinada por el gráﬁco de la función $f$, las rectas $x = −1, x = 1$ y el eje es un semicírculo de radio 1, el área de la región acotada es $||f||_1 = \frac{\pi}{2}.$

$\therefore ||f||_{1} = \frac{\pi}{2}$


### Ejemplo 2.4 
En el espacio $C_{[0,2]}$, para la funcián

$$ f(x)=\left\{ \begin{array}\\ x, & para\ x \ en \ [0,1]; \\ 2-x,  & para \ x \ en \ [1,2]. \\  \end{array}\right. $$

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0,1, num=3)
x2 = np.linspace(1,2, num=1000)

plt.plot(x1, [x for x in x1], 'g', label='f(x)=x, x en [0,1]')
plt.plot(x2, [2-x for x in x2], 'b', label='f(x)=2-x, x en [1,2]')

plt.legend(bbox_to_anchor=(0.,1.2,1., .102), loc='lower left',ncol=2, mode="expand", borderaxespad=0.)

plt.grid()
plt.show()
```

La regián determinada por el gráfico de la función $f$, las rectas $x = 0, x = 2$ y el eje $Ox$ es un triángulo de
base 2 y altura 1. El área de la región acotada es $\frac{(2)(1)}{1} = 1$

$\therefore ||f||_{1} = 1$

### Ejemplo 2.5 

Para la función $f(x)=sen(x)$ del espacio $C_{[0,\pi]}$, se tiene que la norma $\parallel f \parallel _1 = \frac{\pi}{2}$

El área de la región indicada viene dada por la integral 

$$\begin{equation}
\int_{0}^{\pi} sen(x) \ dx
\end{equation}$$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.linspace(-0, np.pi, 200)

plt.axhline(0, color="red")
plt.axvline(0, color="red")


plt.plot(x, np.sin(x), label='f(x)')
plt.title("grafico de la función")
plt.legend()
plt.show()
```

Por lo tanto $\parallel f \parallel _1 =2$ 
La norma $\parallel \cdot \parallel _1$ genera en $C_{[a,b]}$ la siguiente métrica:

$$ d_1 (f,g)= \ \parallel f-g \parallel _1 = \begin{equation} \int_{a}^{b} | f(x)-g(x)| \ dx \end{equation}$$

La métrica $d_1$ representa el área de la región acotada por los gráficos de las funciones $f$ y $g$ y por las rectas $x=a$ y $x=b$. 

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.linspace(-0, np.pi, 200)
y = np.linspace(-0, np.pi, 200)
plt.axhline(0, color="red")
plt.axvline(0, color="red")


plt.plot(x, np.sin(x), label='f(x)')
plt.plot(y, (x-(np.pi)/2)**2+0.7, label='g(x)')
plt.title("grafico de la función")
plt.legend()
plt.show()
```

### Ejemplo 2.6

Para las funciones $f(x)=2$ y $g(x)=3$ del espacio $C_{[-1,1]}$, se tiene que la distancia $d_1 (f,g)=6$

La región acotada por los gráficos de las funciones $f$ y $g$, es un rectángulo de base $2$ y altura $1$, el área de la región es $2$: $d_1 (f,g)=2$

```python
def func(x):
    return x
x = np.linspace(-2, 2)
y = func(x)

plt.hlines(y=1, xmin=0-1, xmax=1, colors='green',ls='-')
plt.hlines(y=2, xmin=-1, xmax=1, colors='green',ls='-')
plt.axhline(0, color="red")
plt.axvline(0, color="red")
```

### Ejemplo 2.7 

Para una función $f$ del espacio $C_{[a,b]}$ y un número $\xi > 0$, la distancia $d_1 (f,f+ \xi ) = \xi \cdot (b-a)$. 

Área de la región: 

$$\begin{equation}
\int_{a}^{b} |f(x) + \xi - f(x)| \ dx
\end{equation}$$

Luego, $d_1 (f, f+ \xi) = \xi \cdot (b-a)$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.linspace(1, 2.9, 200)
y = np.linspace(1, 2.9, 200)
plt.axhline(0, color="red")
plt.axvline(0, color="red")


plt.plot(x, np.e**(-((x-1.5)**2/0.5))+0.5, label='f(x)')
plt.plot(y, np.e**(-((x-1.5)**2/0.5))+0.2, label='g(x)')
plt.title("grafico de la función")
plt.legend()
plt.show()
```

### Ejemplo 2.8 

Para las funciones $f(x) = \sqrt{ 1-x^2 }$ y $g(x) = - \sqrt{1-x^2}$ del espacio $C_{[-1,1]}$, se tiene la distancia $d_1(f,g)= \pi$. 

La región determinada por los gráficos de las dos funcione es un círculo de radio $1$, luego el área es $d_1 (f,g)= \pi$

```python
import numpy as np
from matplotlib import pyplot as plt

num_segmentos = 20
rad = 1
cx = 0
cy = 0

angulo = np.linspace(0, 2*np.pi, 200)
x = rad * np.cos(angulo) + cx
y = rad * np.sin(angulo) + cy

plt.plot(x, y, color="red", markersize=1)
plt.axhline(0, color="blue")
plt.axvline(0, color="blue")

plt.title("Circulos")
plt.xlabel("X")
plt.ylabel("Y")
plt.gca().set_aspect('equal')
plt.show()
```
