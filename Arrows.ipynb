h) $v_1 = (3,1,2)$, $v_2 = (-1,2,4) $$\\$
$T(v_1) = (6,7,4)$, $T(v_2)= (5,-7,8)$

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
soa = np.array([[0, 0, 0, 3, 1, 2], [0, 0, 0, -1, 2, 4]])

X, Y, Z, U, V, W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 3])
ax.set_zlim([-1, 5])
plt.show()
```
