## Ahora vamos a graficar! ##

Vamos a graficar una funcion, lo primero que tenemos que hacer es definirla, para ello utilizaremos el comando def de la siguiente manera:

```python
def f(x):
    return np.sin(x)
g= lambda x: np.cos(x)
```

```python
x= np.linspace(0,20,100)
```

```python
plt.plot(x,f(x),label="seno")
plt.xlabel ("x")
plt.ylabel("y")
plt.title("Función Seno")
plt.plot(x,g(x),label="coseno")
plt.grid()
plt.legend()

            
```
