3.- Encontrar una base y la dimensión del subespacio vectorial

$ S =< (1, 2,−1, 3), (2, 1, 0,−2), (0, 1, 2, 1), (3, 4, 1, 2) >. $

### Solución:

Primero, debemos asegurarnos que el conjunto generador consista en vectores linealmente independientes, para esto, bla bla bla...

Podemos deducir que el conjunto generador del subespacio vectorial puede ser denotado como: $ N = \{ (1,2,−1,3), (2,1,0,−2), (0,1,2,1), (3,4,1,2) \} $, sin embargo, $N$ no es libre, ya que:

$$ (0,0,0,0) = a (1,2,-1,3) + b (2,1,0,-2) + c (0,1,2,1) + d (3,4,1,2) $$

Al solucionar el problema con python encontramos que:

```python
import numpy as np
```

```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import Eq, solve_linear_system, Matrix
from numpy import linalg
import sympy as sp
```

```python
eq1 = sp.Function('eq1')
eq2 = sp.Function('eq2')
eq3 = sp.Function('eq3')
eq4 = sp.Function('eq4')

x,y,z,w = sp.symbols('x y z w')

eq1 = Eq(x+ 2*y -z +3*w, 0)
eq2 = Eq(2*x +y -2*w, 0)
eq3 = Eq(y +2*z +w,0)
eq4 = Eq(3*x +4*y +z +2*w,0)


display(eq1)
display(eq2)
display(eq3)
display(eq4)
```

```python
system = Matrix([[1,2,-1,3],[2,1,0,-2],[0,1,2,1],[3,4,1,2]])
display(system)
display(system.shape)
system.det()
```

```python
display(solve_linear_system(system,x,y,z,w))
```
