## 1. La norma $||f||_\infty $

Denotaremos por $C_{[a,b]}$ el espacio de las funciones continuas  

$C_{[a,b]}=\{f:[a,b]\rightarrow \mathbb{R}: f $ es una función continua $\}$

La función:

$$ ||f||_{\infty} = sup\{ |f(x)| :x \in [a,b] \} $$

es una norma del espacio $C_{[a,b]}$, llamada la norma del supremo.

**Ejemplo 1.1**  En el espacio ($C_{[\frac{1}{2} ,2]}, || · ||_{\infty}$) consideramos la hipérbola básica $f(t) = \frac{1}{t}$

```python
def func(x):
    return 1/x

x = np.linspace(0.5, 2)
y = func(x)

# # Sombrear el area entre y0 y y1
# plt.fill_between(x, y, 0,
#                  facecolor="blue", # The fill color
#                  color='orange',       # The outline color
#                  alpha=0.2)          # Transparency of the fill


plt.plot(x,y, color='black')
plt.vlines(x=0.5, ymin=0, ymax=2, colors='green', ls='--', lw=2)
plt.vlines(x=2, ymin=0, ymax=0.5, colors='green', ls='--', lw=2)
plt.hlines(y=0.5, xmin=0, xmax=2, colors='green',ls='--')
plt.hlines(y=2, xmin=0, xmax=0.5, colors='green',ls='--')
```

La hipérbola es una función decreciente, su valor máximo lo realiza en el extremo $x = \frac{1}{2} = 0,5$.

Por lo tanto $||f||_{\infty} = f( \frac{1}{2} ) = 2$

**Ejemplo 1.2**  En el espacio ($ C_{[0,1]}, || · ||_{\infty} $) consideramos una función que dependa de un parámetro $f(t) = \frac{nt}{n+t}$

La derivada de la función $f$:

$f'(t)=\frac{n^2}{(n+t)^2}>0$ para todo 0$\leq t \leq 1$

La norma $|| · ||_{\infty}$ del espacio $ C_{[0,1]}$ genera en el mismo conjunto la siguiente metrica:

$$ d_{\infty}(f,g) = || f-g ||_{\infty}=sup\{ |f(x)-g(x)| :x \in [a,b] \} $$

```python
def func(x):
    return np.sin(x) + 1

def f2(x):
    return 0.2*(x-1.57)**2 -1

x = np.linspace(0.5, 6)
y1 = func(x)
y2 = f2(x)

plt.plot(x,y1, color='black')
plt.plot(x,y2, color='black')

plt.vlines(x=0.5, ymin=-1, ymax=1.5, colors='green', ls='--', lw=2)
plt.vlines(x=6, ymin=-1, ymax=2.925, colors='green', ls='--', lw=2)
plt.text(1.2, 0.4, 'd∞ (f,g)')
plt.arrow(1.57, f2(1.57), 0, 3, ls='--')


plt.show()
```

**Ejemplo 1.3**  En el espacio ($ C_{[0,1]}, d_{\infty} $) consideramos las siguientes funciones

$f(t) = t$, $g(t)=t^2$, $h(t)=t^3$ y $q(t)=0$, para $0\leq t \leq1$

Determinamos las siguientes distancias:

$d_{\infty}(f,g),d_{\infty}(g,h)$ y $d_{\infty}(h,q)$

```python
#Definimos la función
def f(x,n):
    if x>=0 and x<=1:
        return x**n
    elif x>1: 
        return x
```

```python
x=np.linspace(0,1,100)

f1=[f(i,1) for i in x]
f2=[f(i,2) for i in x]
f3=[f(i,3) for i in x]
q4=plt.plot([0,0])

plt.plot(x,f1,label="f(t)")
plt.plot(x,f2,label="g(t)")
plt.plot(x,f3,label="h(t)")

plt.hlines(y=1, xmin=0, xmax=1, colors='black',ls='--')
plt.vlines(x=1, ymin=0, ymax=1, colors='black', ls='--', lw=2)
plt.vlines(x=0, ymin=0, ymax=1, colors='black', ls='-', lw=2)

plt.legend()
plt.grid()
plt.title("Intervalo [0,1]")
```

El nuumero $d_\infty(f, g)$ mide la mayor abertura entre los graficos de $f$ y $g$.Para calcularlo debemos determinar el maximo de $\varphi(t)$ $=| t−t^2 |$ $= t−t^2$ en el intervalo $[0, 1]$. La derivada $\varphi(t)$ $= 1−2t$ se anula en el punto t = $\frac{1}{2}$ . Luego,el maximo valor de la funcion $\varphi$ es el numero $\varphi(\frac{1}{2})= \frac{1}{4}$, asi $d_\infty(f, g) = \frac{1}{4}$.


**Ejemplo 1.4** Para la funcion $f$ del espacio $ C_{[0,1]} $ y cada $\xi>0$ se cumple que: $d_{\infty}(f,f+\xi)=\xi$



La funcion $f +\xi$  corresponde a subir el grafico de $f $una cantidad de $\xi$ unidades. Luego, la distancia entre los graficos es exactamente $\xi$.

Por lo tanto $d_\infty(f,f+\xi)=\xi$

```python
def h(x):
    return 2*np.cos(x-1)+3
def g(x):
    return 2*np.cos(x-1)-1/2
x=np.linspace(-4,6,100)

plt.plot(x,h(x),label="hola")
plt.plot(x,g(x),label="hola2")
plt.vlines(x=0, ymin=-3, ymax=5, colors='black', ls='-', lw=2)
plt.hlines(y=0, xmin=-4, xmax=6, colors='black',ls='-')
plt.vlines(x=6, ymin=0, ymax=4, colors='black', ls='--', lw=2)
plt.vlines(x=-4, ymin=0, ymax=4, colors='black', ls='--', lw=2)
plt.grid()
```
