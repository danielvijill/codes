## 2. Demuestra que los siguientes subconjuntos de $R^3$ son subespacios vectoriales

#### a). El conjunto de vectores $(x,y,z)$ tales que $x+y+z=0$

##### Solución

El subconjunto al que llamaremos $S$ define un plano que contiene al vector $(0,0,0)$, satisface la primera condición para un subespacio vectorial:
$$x+y+z=0$$

$$0+0+0=0$$

$$0=0$$

Supongamos que: $(x,y,z)ϵS$ ^ $(x',y',z')ϵS$

Cerradura para la suma: $$(x,y,z)+(x',y',z')ϵS$$ $$(x+x',y+y',z+z')ϵS$$
Tenemos la condición $x+y+z=0$, por lo que: $$(x+x')+(y+y')+(z+z')=0$$ $$(x+y+z)+(x'+y'+z')=0$$ $$0+0=0$$
$$0=0$$
Se satisface la Cerradura para la suma.

Ahora comprobaremos la Cerradura para el producto: $$α(x,y,z)ϵS$$ $$(αx+αy+αz)ϵS$$ 
Aplicando la condición $x+y+z=0$ se tiene que: $$α(x+y+z)=0$$ $$α(0)=0$$ $$0=0$$ Y se satisface la cerradura para el producto.

Concluimos pues, que $S$ es un Subespacio Vectorial.


#### b). El conjunto de vectores $(x,y,z)$ tales que $x=y$ y $2y=z$

##### Solución

Llamaremos $S$ al subcojunto. Podemos comprobar que $S$ contiene al vector $(0,0,0)$, satisface la primera condición para un subespacio vectorial:
$$x-y=0$$ $$2y-z=0$$ 


Supongamos que: $(x,y,z)ϵS$ ^ $(x',y',z')ϵS$

Cerradura para la suma: $$(x,y,z)+(x',y',z')ϵS$$ $$(x+x',y+y',z+z')ϵS$$

Tenemos las condiciones de $S$: $$x-y=0$$ y $$2y-z=0$$

Por lo que: $$(x+x')-(y+y')=0$$ $$(x-y)+(x'-y')=0$$ $$0+0=0$$
$$0=0$$
y también:
$$2(y+y')-(z+z')=0$$ $$(2y-z)+(2y'-z')=0$$ $$0+0=0$$ $$0=0$$
Se satisface la Cerradura para la suma.

Ahora comprobaremos la Cerradura para el producto: $$α(x,y,z)ϵS$$ $$(αx+αy+αz)ϵS$$
Aplicando las condiciones se tiene que: $$α(x-y)=0$$ $$α(0)=0$$ $$0=0$$
y también $$α(2y-z)=0$$ $$α(0)=0$$ $$0=0$$
Y se satisface la cerradura para el producto.

Por lo tanto, $S$ es un Subespacio Vectorial.

#### c). El conjunto de vectores $(x,y,z)$ tales que $x+y=3z$

##### Solución

De igual manera, llamaremos $S$ al subconjunto. Escribiremos la condición del conjunto como $x+y-3z=0$. Este también define un plano que contiene al vector $(0,0,0)$, satisface la primera condición para un subespacio vectorial:
$$x+y-3z=0$$ $$0+0+0=0$$ $$0=0$$

Supongamos que: $(x,y,z)ϵS$ ^ $(x',y',z')ϵS$

Cerradura para la suma: $$(x,y,z)+(x',y',z')ϵS$$ $$(x+x',y+y',z+z')ϵS$$
Tenemos la condición $x+y-3z=0$, por lo que: $$(x+x')+(y+y')-3(z+z')=0$$ $$(x+y-3z)+(x'+y'-3z')=0$$ $$0+0=0$$
$$0=0$$
Se satisface la Cerradura para la suma.

Ahora comprobaremos la Cerradura para el producto: $$α(x,y,z)ϵS$$ $$(αx+αy+αz)ϵS$$ 
Aplicando la condición $x+y-3z=0$ se tiene que: $$α(x+y-3z)=0$$ $$α(0)=0$$ $$0=0$$ Y se satisface la cerradura para el producto.

Por lo tanto, $S$ es un Subespacio Vectorial.
