## 2. Aproximación de una función mediante polinomios ortogonales
Aunque todavía no hemos estudiado con detalle, ya se ha introducido una familia de
polinomios otogonales: Los polinomios de Legendre. Usando estos polinomios en el
intevalo apropiado, realice una aproximación de las funciones aproximadas en la
sección anterior.

$ f_1(x)=1 \\ $
$ f_2(x)=x \\ $
$f_3(x)=x^2 \\ $
$f_4(x)=cos(\frac{1}{2} \pi x) $

$f_5(x)= Sin(x) \\ $
$f_6(x)= e^x \\$
$f_7(x)= \frac{1}{1+25x^2} \\ $

```python
x = np.linspace(1,10,1000)
f0 = x/x
```

```python
degrees = [0, 1, 2, 3]
plt.plot(x, f0, c='k', lw=1, label='f0')
for deg in degrees:
    #coeffs = myfitfun(x, f0, deg)
    coeffs = np.polynomial.legendre.legfit(x, f0, deg)
    plt.plot(x, np.polynomial.legendre.legval(x, coeffs),linewidth=2, label="Grado=%i"%deg)
plt.legend()
```
