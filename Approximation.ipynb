```python
from sympy import init_printing
init_printing(use_latex=True)
from sympy import *
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import pyplot
import ipywidgets as wg
from IPython.display import display
import numpy as np
from sympy.abc import x
from sympy import symbols
from sympy.plotting import plot
```

```python
def f(x):
    return cos(pi*x/2)
def g(x):
    return E**x

def c(j):
    return integrate(f(x)*x**j, (x,a,b))
```

```python
def Fit(func,x):
    n=10
    i=1
    while i<= n:
        k = i #int(input("ingrese el número de columnas: "))
        j = i #int(input("ingrese el número de filas: "))
        a = 1 #int(input("ingrese el límite inferior del intervalo: "))
        b = -1 #int(input("ingrese el límite superior del intervalo: "))
        def M(k,j):
            celda=(b**(k+j+1)-(a)**(k+j+1))/(k+j+1)
            return celda

        k1 = i #int(input("ingrese el número de filas: "))
        j1 = 1 # int(input("ingrese el número de columnas: "))
        a1 = 1 #int(input("ingrese el límite inferior del intervalo: "))
        b1 = -1 #int(input("ingrese el límite superior del intervalo: "))
        def C1(k1,j1):
            celda=(integrate(func*x**k1, (x,a1,b1)))
            return celda
        W=Matrix(k1, j1, C1)

        B=Matrix(k, j, M)

        H=np.array(B)
        H1=np . array ( H ,  dtype = float )
        H2=np.array(W)
        H3=np . array ( H2 ,  dtype = float )

        G=np.linalg.inv(H1).dot(H3)

        Q=0
        j=0
        for elemento in np.nditer(G,order='F'):
            Q+=elemento*x**j
            j+=1

        p1 = plot(Q,(x,-1,1),axis_center=(0,0),line_color='r', show=False)
        p2 = plot(func,(x,-1,1), show=False)
        p1.append(p2[0])
        p1.show()

        i=i+1
```

```python
Fit(f(x),x)
```
