## 3. La Norma $ ||f||_2 = \sqrt{\int_a^b |f(x)|^2 dx} $

Otra norma del espacio de las funciones continuas $C_{[a,b]}$ es la función:
$$ ||f||_2 = \sqrt{\int_a^b |f(x)|^2 dx} $$

**Ejemplo 3.1** Para la función $f(x) = \sqrt{r^2 -x^2}$ del espacio $C_{[−r,r]}$, se tiene que la norma $||f||_2 = \frac{2r\sqrt{r}}{\sqrt{3}} $.

Cuando giramos la región que determina esta función $f$, alrededor del eje, se genera la esfera de radio $r$. El volumen de una esfera de radio r está dado por el número $\frac{4\pi r^3}{3}$. Luego,
$$ \pi ||f||_2^2 = \frac{4\pi r^3}{3} $$

```python
def f(x,r):
    return np.sqrt(r**2 -(x**2))

a, zero, b = -2, 0, 2
x = np.linspace(-2, 2, 100)

y = f(x,2)
y2 = (-1)*f(x,2)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.set_ylim(bottom=0)

plt.fill_between(x,y,0, alpha=0.2)

# plt.hlines(y=0, xmin=-3, xmax=3, color='r', ls='--')

plt.gca().set_aspect('equal', adjustable='box')

# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)
# ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, zero, b))
ax.set_xticklabels(('$-r$', '0', '$r$'))
ax.set_yticks([])


fig.text(1, 0.15, '$x$')
fig.text(0.11, 0.9, '$y$')

plt.plot(x,y)
plt.plot(x,y2)
```

</div>

Por lo tanto, $||f||_2 = \frac{2r\sqrt{r}}{\sqrt{3}} $

**Ejemplo 3.2** Para $r > 0$ y la función constante $f(x) = r$ del espacio $C_{[0,h]}$, se tiene que la norma $||f||_2 = r ·\sqrt{h}$.

Cuando giramos la región alrededor del eje de las abscisas se genera el cilindro circular recto de radio $r$ y altura $h$. El volumen del cilindro está dado por el número $\pi r^2 h$. Luego, 
$$ \pi||f||_2^2 = \pi r^2 h $$

```python
a, b = 0, 1
x = np.linspace(0, 1, 100)
y=np.linspace(1,1,100)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.set_ylim(bottom=0)

plt.fill_between(x,y,0, alpha=0.2)

# plt.hlines(y=0, xmin=-1, xmax=3, color='black') 

plt.gca().set_aspect('equal', adjustable='box')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, b))
ax.set_xticklabels(('$0$', '$r$'))
ax.set_yticks([])

plt.plot(x,y)
```

Rotando la figura obtenemos:

<div>

</div>

Por lo tanto, $||f||_2 = r \cdot \sqrt{h} $

**Ejemplo 3.3** Para la función $f(x) = x$ del espacio $C_{[0,r]}$, se tiene que la norma $||f||_2 = \frac{r\sqrt{r}}{\sqrt{3}} $.

Cuando giramos la región alrededor del eje de las abscisas se genera el cono circular recto de radio $r$ y altura $r$. El volumen del cono está dado por el número $\pi \frac{r^3}{3}$. Luego,
$$ \pi||f||_2^2 = \pi \frac{r^3}{3} $$

```python
def f(x):
    return (r*x)

a, b = 0, 2
x = np.linspace(0, 2, 100)
r=1
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.set_ylim(bottom=0)

plt.fill_between(x,y,0, alpha=0.2)

# plt.hlines(y=0, xmin=-3, xmax=3, color='r', ls='--')

plt.gca().set_aspect('equal', adjustable='box')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, b))
ax.set_xticklabels(('$0$', '$r$'))
ax.set_yticks([])

plt.plot(x,y)
```

Rotando la figura obtenemos:

<div>

</div>

Por lo tanto, $||f||_2 = \frac{r\sqrt{r}}{\sqrt{3}} $

La norma $|| · ||_2$ genera la siguiente distancia en el espacio $C_{[a,b]}$:

$$
d_{2}(f, g)=\|f-g\|_{2}=\sqrt{\int_{a}^{b}|f(x)-g(x)| d x}
$$

De manera similar a la norma, la métrica cumple con la propiedad de que el número
$$ \pi \cdot d_{2}(f, g)^{2} $$

es el volumen del sólido que se genera al girar la región determinada por las funciones $f$ y $g$ alrededor del eje de las abscisas.

**Ejemplo 3.4** Para las funciones constantes $f(x) = 1$ y $g(x) = 2 $ del espacio
$C_{[0,h]}$, se tiene que la distancia $d_2(f, g) = \sqrt3 h$.

Cuando giramos la región determinada por las funciones alrededor del eje de las abscisas se genera el cono circular recto hueco, de radio externo 2, radio interno 1 y altura h. El volumen de este cilindro viene dado por el número $3\pi h$. Luego,
$$
\pi d_{2}^{2}(f, g)=3 \pi h
$$

```python
a, b = 0, 1
x = np.linspace(0, 1, 100)
y = np.linspace(1,1,100)
y2 = np.linspace(2,2,100)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.plot(x, y2, 'r', linewidth=2)
ax.set_ylim(bottom=0)

plt.fill_between(x,y,y2, alpha=0.2)

# plt.hlines(y=0, xmin=-1, xmax=3, color='black') 

plt.gca().set_aspect('equal', adjustable='box')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, b))
ax.set_xticklabels(('$0$', '$r$'))
ax.set_yticks([])

# plt.plot(x,y)
plt.plot(x,y2)
```

<div>

</div>

Luego, $d_{2}(f, g)=\sqrt{3} h$
