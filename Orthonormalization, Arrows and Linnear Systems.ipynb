El producto interno para polinomios esta definido como:
$$ \langle p, q \rangle = \int_0^1 p(x)q(x)\ dx,
 $$

$ \langle 1, x \rangle = \int_0^1 x dx = \frac{1}{2} $

$ \langle 1, x^2 \rangle = \int_0^1 x^2 dx = \frac{1}{3} $

A continuación utilizamos el proceso de Gram Schmidt para ortonormalizar la base $B_2$

```python
from sympy import init_printing
init_printing(use_latex=True)
```

```python
x0, x1, t = sp.symbols('x0 , x1, t')

def Inner(x0,x1):
    ip = sp.integrate( x0 * x1, (t, 0, 1))
    return ip
```

```python
display(Inner( t*t, t*t) )
```

Ahora que la función de producto interno está construída,

```python
a0,a1,a2 = sp.symbols('a0,a1,a2')
e0,e1,e2 = sp.symbols('e0,e1,e2')
e1p,e2p = sp.symbols('e1p,e2p')

def Gram(a0,a1,a2):
    e0 = a0/ sp.sqrt(Inner(a0,a0))
    
    e1p = a1 - Inner(a1,e0) * e0         #ortogonalizacion
    e1 = e1p / sp.sqrt(Inner(e1p, e1p))  #normalizacion
    
    e2p = a2  - ( Inner(a2, e1) * e1 ) - ( Inner(a2,e0) * e0 ) #ortogonalizacion
    e2 = e2p / sp.sqrt(Inner(e2p, e2p))                        #normalizacion
    
    return e0, e1, e2
```

Una vez creada la función del proceso de ortonormalización, introducimos nuestros vectores $x(0), x(1)$ y $x(2)$

```python
Gram(1, t, t*t)
```

Almacenamos estos resultados en un arreglo para su futura utilización

```python
el = [Gram(1, t, t*t)]
Elements = np.array(el)
```

Finalmente, los vectores ortonormalizados son:

```python
display(Gram(1, t, t*t))
```

Para demostrar que los vectores son, en efecto ortogonales y normales, podemos realizar las siguientes pruebas con el producto interno:

Ortogonalidad:

```python
display(Inner(Elements[0][0],Elements[0][1]))
display(Inner(Elements[0][1],Elements[0][2]))
display(Inner(Elements[0][2],Elements[0][0]))
```

Normalidad

```python
display(Inner(Elements[0][0],Elements[0][0]))
display(Inner(Elements[0][1],Elements[0][1]))
display(Inner(Elements[0][2],Elements[0][2]))
```
