# Ejercicio 2.19 
Demuestre que $T: R^2 \rightarrow R^3$, dado por $T(x,y) = (x^2+y^2,x+y,2x-y)$ no es una transformación lineal

```python
import sympy as sp
import numpy as np
from numpy import *
from sympy import *
init_printing(use_latex=True)
```

Primero debemos crear la matriz que compone la función de la transfromada lineal

```python
x, y, z, t = symbols('x, y, z, t')
A = Matrix([[x*x+y*y],[x+y],[2*x-y]])
display(A)
```

Ahora comprobamos las 2 propiedades necesarias para demostrar que $T(x,y) = (x^2+y^2,x+y,2x-y)$ es una transformada lineal:

$$T(u+v)=T(u)+T(v)$$

$$T( a k)=a T(k)$$

Siendo $u=(x_1,y_1), v=(x_2,y_2)  , k=(x,y), a \in \mathbb{R} $

## 1. $T(u+v)=T(u)+T(v)$

### $T(u+v)$

```python
x_1, y_1, x_2, y_2 = symbols('x_1, y_1, x_2, y_2')
A = Matrix([[(x_1+y_1)*(x_1+y_1)+(x_2+y_2)*(x_2+y_2)],[(x_1+y_1)+(x_2+y_2)],[2*(x_1+y_1)-(x_2+y_2)]])
display(A)
```

Lo que es igual a:

```python
x_1, y_1, x_2, y_2 = symbols('x_1, y_1, x_2, y_2')
A = Matrix([[x_1*x_1+2*x_1*y_1+y_1*y_1+x_2*x_2+2*x_2*y_2+y_2*y_2],[(x_1+y_1)+(x_2+y_2)],[2*(x_1+y_1)-(x_2+y_2)]])
display(A)
```

### $T(u)+T(v)$

```python
x_1, y_1, x_2, y_2 = symbols('x_1, y_1, x_2, y_2')
A = Matrix([[x_1*x_1+y_1*y_1],[x_1+y_1],[2*x_1-y_1]])
B = Matrix([[x_2*x_2+y_2*y_2],[x_2+y_2],[2*x_2-y_2]])
C = A+B
display(C)
```

Analizando las matrices podemos decir que $T(u+v) \neq T(u)+T(v)$

Nota: podemos afirmar que $T(u+v)$ no es una transfrmación lineal, pero seguiremos con la segunda propiedad de forma demostrativa.

## 1. $T( a k)= a T(k)$

### $T( a k)$

```python
x, y, a = symbols('x, y, a')
A = Matrix([[(a*x)*(a*x)+(a*y)*(a*y)],[a*x+a*y],[2*a*x-a*y]])
display(A)
```

### $aT(k)$

```python
x, y, a = symbols('x, y, a')
A = a*Matrix([[x*x+y*y],[x+y],[2*x-y]])
display(A)
```

```python
x, y, a = symbols('x, y, a')
A = Matrix([[a*x*x+a*y*y],[a*x+a*y],[2*a*x-a*y]])
display(A)
```

Analizando las matrices podemos decir que $T(ak) \neq aT(k)$

## Conclusión 

Debido que ninguna de las propiedades se demostró, podemos decir que $T(x,y) = (x^2+y^2,x+y,2x-y)$  no es una transformación lineal.

\end{array}\right)
$$

Para comprobar que la matriz es es unitaria, el producto de su transpuesta conjugada con ella misma debe ser igual a la matriz identidad.

```python
import sympy as sp
from sympy import *
from sympy.physics.quantum.dagger import Dagger
init_printing(use_unicode=True)
```

```python
A = Matrix([[1/sqrt(2),-I/sqrt(2),0],[1/sqrt(6),I/sqrt(6),-2/sqrt(6)],[(1+I)/sqrt(6),(-1+I)/sqrt(6),(1+I)/sqrt(6)]])
display(A)
```

```python
B = Dagger(A)
display(B)
    
```

```python
C = sp.simplify(A*B)
display(C)
```

Este resultado, muestra la matriz identidad. Por lo tanto, la matriz de transformación es unitaria.

# Espacio Dual
Ejercicio 2. Dada la base $B$ del **K-espacio vectorial $V$** , hallar su base dual en cada uno de
los siguientes casos:

### i) $V = \mathbb{R}^2$, $B = \{ (-1,1),(2,0) \}$

Sea $\beta ^* = \{ \varphi_1, ..., \varphi_n \}$ la base dual del espacio $B$ generado por los vectores $B = \{ v_1, .., v_n \}$

Siguiendo la definición de base dual, sabemos que:
$\varphi_i(v_j)$ esta definido como:
$$
  \begin{cases}
                                   0 & \text{si $i \neq j$} \\
                                   1 & \text{si $i=j$} \\
  \end{cases}
 \tag 1$$

Comenzamos por expresar el vector $\varphi_n$ en terminos de la base canónica en $\mathbb{R}^2$
$$ \varphi_n(x,y) = \varphi_n(x*(1,0)+y*(0,1)) $$
$$ \varphi_n(x,y) = x * \varphi_n (1,0)+ y * \varphi_n (0,1) \tag 2$$

Mediante esta expresión, trabajamos con los puntos $(x,y)$ para los cuales sabemos el valor de la función:

$ \varphi_1 (-1,1) = 1 $   $\implies$    $ \varphi_1(x,y) = - \varphi_1 (1,0)+ \varphi_1 (0,1) = 1 $

$ \varphi_1 (2,0) = 0 \implies \varphi_1(x,y) = 2 \varphi_1 (1,0) + 0\varphi_1 (0,1) = 0 $

Resolvemos el sistema de ecuaciones resultante

```python
import numpy as np
from numpy import *
```

```python
A = np.array([[-1,1],[2,0]])
b = np.array([1,0])
x = np.linalg.solve(A, b)

display(x)
```

El resultado del sistema es $\varphi_1 (1,0) = 1 $ y $\varphi_1 (0,1) = 0 $, retomando la expresión en $(2)$, tenemos que:
$$ \varphi_1(x,y) = x * \varphi_1 (1,0)+ y * \varphi_1 (0,1) $$
$$ \varphi_1(x,y) = x (1) + y (0) $$
$$ \varphi_1(x,y) = x \tag 3$$

De manera similar, repetimos el proceso anterior para $\varphi_2$ y encontraremos que:

```python
A = np.array([[-1,1],[2,0]])
b = np.array([0,1])
x = np.linalg.solve(A, b)

display(x)
```

El resultado del sistema es $\varphi_2 (1,0) = \frac{1}{2} $ y $\varphi_2 (0,1) = \frac{1}{2} $, retomando la expresión en $(2)$, tenemos que:
$$ \varphi_2(x,y) = x * \varphi_2 (1,0)+ y * \varphi_2 (0,1) $$
$$ \varphi_2(x,y) = x (\frac{1}{2}) + y (\frac{1}{2}) $$
$$ \varphi_2(x,y) = \frac{x+y}{2} \tag 4$$

Y de esa manera, la base del espacio dual $B^*$ es:
$$ \beta ^* = \left\{ x , \frac{x+y}{2} \right\} $$

## iii) $ V = \mathbb{R}_3[X]$, $ B = \{ -X+2 , X-1 , X^2 -3X+ 2, X^3 -3X^2 +2X \} $

Comenzamos por expresar el vector $\varphi_n$ en terminos de la base canónica en $\mathbb{R}_n$
$$ \varphi_n(X^3,X^2,X,1) = \varphi_n(X^3 *(1,0,0,0)+X^2 *(0,1,0,0)+ X*(0,0,1,0)+ 1*(0,0,0,1)) $$
$$ \varphi_n(X^3,X^2,X,1) = X^3 * \varphi_n (1,0,0,0)+ X^2 * \varphi_n (0,1,0,0)+ X * \varphi_n (0,0,1,0) + 1 * \varphi_n (0,0,0,1) $$

**Para $\varphi_1$:**

$ \varphi_1 (X^3 -3X^2 +2X) = 1 $   $\implies$    $ \varphi_1(X^3,X^2,X,1) =  X^3 * \varphi_1 (1,0,0,0)-3 X^2 * \varphi_1 (0,1,0,0)+ 2X * \varphi_1 (0,0,1,0) $
