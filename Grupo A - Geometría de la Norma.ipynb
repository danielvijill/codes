## 1. La norma $||f||_\infty $

Denotaremos por $C_{[a,b]}$ el espacio de las funciones continuas  

$C_{[a,b]}=\{f:[a,b]\rightarrow \mathbb{R}: f $ es una funcin continua $\}$

La funcin:

$$ ||f||_{\infty} = sup\{ |f(x)| :x \in [a,b] \} $$

es una norma del espacio $C_{[a,b]}$, llamada la norma del supremo.

### Ejemplo 1.1
En el espacio ($C_{[\frac{1}{2} ,2]}, ||  ||_{\infty}$) consideramos la hiprbola bsica $f(t) = \frac{1}{t}$

```python
def func(x):
    return 1/x

x = np.linspace(0.5, 2)
y = func(x)

# Sombrear el area entre y0 y y1
# plt.fill_between(x, y, 0,
#                  facecolor="blue", # The fill color
#                  color='orange',       # The outline color
#                  alpha=0.2)          # Transparency of the fill


plt.plot(x,y, color='black')
plt.vlines(x=0.5, ymin=0, ymax=2, colors='green', ls='--', lw=2)
plt.vlines(x=2, ymin=0, ymax=0.5, colors='green', ls='--', lw=2)
plt.hlines(y=0.5, xmin=0, xmax=2, colors='green',ls='--')
plt.hlines(y=2, xmin=0, xmax=0.5, colors='green',ls='--')
```

La hiprbola es una funcin decreciente, su valor mximo lo realiza en el extremo $x = \frac{1}{2} = 0,5$.

Por lo tanto $||f||_{\infty} = f( \frac{1}{2} ) = 2$

### Ejemplo 1.2
En el espacio ($ C_{[0,1]}, ||  ||_{\infty} $) consideramos una funcin que dependa de un parmetro $f(t) = \frac{nt}{n+t}$

La derivada de la funcin $f$:

$f'(t)=\frac{n^2}{(n+t)^2}>0$ para todo 0$\leq t \leq 1$

La norma $||  ||_{\infty}$ del espacio $ C_{[0,1]}$ genera en el mismo conjunto la siguiente metrica:

$$ d_{\infty}(f,g) = || f-g ||_{\infty}=sup\{ |f(x)-g(x)| :x \in [a,b] \} $$

```python
def func(x):
    return np.sin(x) + 1

def f2(x):
    return 0.2*(x-1.57)**2 -1

x = np.linspace(0.5, 6)
y1 = func(x)
y2 = f2(x)

plt.plot(x,y1, color='black')
plt.plot(x,y2, color='black')

plt.vlines(x=0.5, ymin=-1, ymax=1.5, colors='green', ls='--', lw=2)
plt.vlines(x=6, ymin=-1, ymax=2.925, colors='green', ls='--', lw=2)
plt.text(1.2, 0.4, 'd (f,g)')
plt.arrow(1.57, f2(1.57), 0, 3, ls='--')


plt.show()
```

$d_{\infty}(f,g),d_{\infty}(g,h)$ y $d_{\infty}(h,q)$

El nuumero $d_\infty(f, g)$ mide la mayor abertura entre los graficos de $f$ y $g$.Para calcularlo debemos determinar el maximo de $\varphi(t)$ $=| tt^2 |$ $= tt^2$ en el intervalo $[0, 1]$. La derivada $\varphi(t)$ $= 12t$ se anula en el punto t = $\frac{1}{2}$ . Luego,el maximo valor de la funcion $\varphi$ es el numero $\varphi(\frac{1}{2})= \frac{1}{4}$, asi $d_\infty(f, g) = \frac{1}{4}$.

La norma $\parallel \cdot \parallel _1$ genera en $C_{[a,b]}$ la siguiente mtrica:

## 3. La Norma $ ||f||_2 = \sqrt{\int_a^b |f(x)|^2 dx} $

Otra norma del espacio de las funciones continuas $C_{[a,b]}$ es la funcin:
$$ ||f||_2 = \sqrt{\int_a^b |f(x)|^2 dx} $$

### Ejemplo 3.1
Para la funcin $f(x) = \sqrt{r^2 -x^2}$ del espacio $C_{[r,r]}$, se tiene que la norma $||f||_2 = \frac{2r\sqrt{r}}{\sqrt{3}} $.

Cuando giramos la regin que determina esta funcin $f$, alrededor del eje, se genera la esfera de radio $r$. El volumen de una esfera de radio r est dado por el nmero $\frac{4\pi r^3}{3}$. Luego,
$$ \pi ||f||_2^2 = \frac{4\pi r^3}{3} $$

```python
def f(x,r):
    return np.sqrt(r**2 -(x**2))

a, zero, b = -2, 0, 2
x = np.linspace(-2, 2, 100)

y = f(x,2)
y2 = (-1)*f(x,2)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.set_ylim(bottom=0)

plt.fill_between(x,y,0, alpha=0.2)

# plt.hlines(y=0, xmin=-3, xmax=3, color='r', ls='--')

plt.gca().set_aspect('equal', adjustable='box')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, zero, b))
ax.set_xticklabels(('$-r$', '0', '$r$'))
ax.set_yticks([])


fig.text(1, 0.15, '$x$')
fig.text(0.11, 0.9, '$y$')

plt.plot(x,y)
plt.plot(x,y2)
```

</div>

Por lo tanto, $||f||_2 = \frac{2r\sqrt{r}}{\sqrt{3}} $

La norma $||  ||_2$ genera la siguiente distancia en el espacio $C_{[a,b]}$:

$$
d_{2}(f, g)=\|f-g\|_{2}=\sqrt{\int_{a}^{b}|f(x)-g(x)| d x}
$$

De manera similar a la norma, la mtrica cumple con la propiedad de que el nmero
$$ \pi \cdot d_{2}(f, g)^{2} $$

es el volumen del slido que se genera al girar la regin determinada por las funciones $f$ y $g$ alrededor del eje de las abscisas.

### Ejemplo 3.4
