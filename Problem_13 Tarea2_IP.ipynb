$ \{ v_1 = (3, 2, 0, 5), v_2 = (-1, 0, 3, -4), v_3 = (2, 2, 3, 1), v_4 = (0, 2, -9, 17) \} $

### Soluci√≥n:

Sea $V$ el espacio vectorial generado por el conjunto de vectores {$ v_1 , ... , v_n $}. Procedemos a verificar si los vectores $\{ v_1 , v_2 , v_3 , v_4 \} $ son linealmente independientes:

```python
import numpy as np
```

```python
A = np.array([[3, 2, 0, 5],[-1, 0, 3, -4],[2, 2, 3, 1],[0, 2, -9, 17]])
b = np.array([0, 0, 0, 0])
#x = np.linalg.solve(A, b)

#x
```

Al analizar el sistema homog√©neo $Ax = b$ formado por la matriz de coeficientes (vectores generadores) $A$, notamos que el sistema no es libre, pues la funci√≥n `np.linalg.solve` retorna un error `LinAlgError: Singular matrix`, lo que nos indica que existe un vector que es combinaci√≥n lineal de otros vectores en el conjunto generador.

$$ (2, 2, 3, 1) = (3, 2, 0, 5) + (-1, 0, 3, -4) $$ y que
$$ (0,2,-9,17) = (3,2,0,5) + 3(-1,0,3,-4) $$

Lo que indica que dos vectores son combinaci√≥nes lineales de otros, y por lo tanto, omitimos dichos vectores en el conjunto, de esa manera, obtenemos que: $ \{ v_1 = (3, 2, 0, 5), v_2 = (-1, 0, 3, -4) \} $ es un conjunto generador del espacio vectorial.
