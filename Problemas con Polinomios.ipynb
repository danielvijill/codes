## 1.- Determinar el valor de $x$ para que el vector $(1, x, 5)  \mathbb{R}^{3} $ pertenezca al subespacio $ <(1, 2, 3), (1, 1, 1) > $.

### Solucin:

Para que el vector $(1,x,5)$ pertenezca a dicho subespacio vectorial, este deber ser representado como una combinacin lineal de los vectores que componen la base del subespacio vectorial, as:

$(1,x,5) = \alpha (1,2,3) + \beta (1,1,1) $

Si y solo si existen $\alpha, \beta \in \mathbb{C} $ tales que dicha condicin se cumple.

Por lo tanto:
$$ 1 = \alpha + \beta $$
$$ x = 2\alpha + \beta $$
$$ 5 = 3\alpha + \beta $$

Utilizamos la herramienta python para solucionar el sistema de 3 ecuaciones y 3 incognitas:

```python
import numpy as np
```

```python
A = np.array([[1,1,0],[2,1,-1],[3,1,0]])
b = np.array([1,0,5])
x = np.linalg.solve(A, b)

print(x)
print(' ')
print('El valor de x que satisface dicha ecuacin es:', int(x[2]) )
```

Como resultado, obtuvimos que $\alpha = 2,$ $ \beta = -1$ y finalmente $x = 3$.

no es libre ya que:

-\alpha_{1}+2\alpha_{3}+\alpha_{4} =0\\

\right.
\end{equation}
