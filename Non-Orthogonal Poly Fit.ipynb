```python
from sympy import init_printing
init_printing(use_latex=True)
from sympy import *
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import pyplot
import ipywidgets as wg
from IPython.display import display
import numpy as np
from sympy.abc import x
from sympy import symbols
from sympy.plotting import plot
```

#### S= Función
#### n= Numero de aproximaciones
#### a=Intervalo inferior
#### b=intervalo superior

```python
def Aproximación(S,n,a,b):
    i=2
    while i<= n:
        k = i
        j = i
        def M(k,j):
            celda=(b**(k+j+1)-(a)**(k+j+1))/(k+j+1)
            return celda
        k1 = i
        j1 =1
        def C1(k1,j1):
            celda=(integrate(S*x**k1, (x,a,b)))
            return celda
        W=Matrix(k1, j1, C1)
        B=Matrix(k, j, M)
        
        H=np.array(B)
        H1=np . array ( H ,  dtype = float )
        H2=np.array(W)
        H3=np . array ( H2 ,  dtype = float )

        G=np.linalg.inv(H1).dot(H3)

        Q=0
        j=0
        for elemento in np.nditer(G,order='F'):
            Q+=elemento*x**j
            j+=1
        display(Q)
        p1 = plot(Q,(x,a,b),axis_center=(0,0),line_color='r', show=False)
        p2 = plot(S,(x,a,b), show=False)
        p1.append(p2[0])
        p1.show()

        i=i+1
```

```python
Aproximación(1,3,-1,1)
```

```python
Aproximación(x,10,-1,1)
```

```python
Aproximación(x**2,10,-1,1)
```
